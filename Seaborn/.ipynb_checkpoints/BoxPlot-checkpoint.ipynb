```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### median, quartile(25,50,75), min-max*1.5, outlier

#### median--> grafikteki yeri:Kutunun icindeki cizgidir.<br>%25 quartile-->grafikteki yeri: kutunun alt kenari<br>%75 quartile-->grafikteki yeri: kutunun ust kenaridir<br>Bıyıklar: Bıyıkların uzunluğu, verinin geri kalanını gösterir. Geleneksel olarak, bıyıklar IQR'ın 1.5 katı kadar uzatılır. Yani, Q1 - 1.5*IQR ve Q3 + 1.5*IQR arasındaki en uç değerlere kadar uzanır.

```python
veri=np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
```

```python
plt.figure(figsize=(6,5)) #ilk deger genislik, ikinci deger ise uzunluk
sns.boxplot(y=veri)
plt.title("Box Plot")
plt.ylabel("data value")
plt.grid(True) #izgara eklemek icin
plt.show()
```

#### seaborn'un kendi datasetleri vardir ve kullanislidir -->__load_dataset()__ kullanarak cagirabiliriz

```python
df = sns.load_dataset("titanic") #titanic datasetini aldik
```
