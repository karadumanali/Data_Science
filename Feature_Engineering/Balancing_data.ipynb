```python
import numpy as np
import pandas as pd
```

##### ya diyelim ki elimizde tahmin icin bir data seti var. kanser testi olsun. Elimizde 1000 data var 900 u kanser degil 100 u kanser.<br>elimize başka bir veri geldigi zaman bu veriyi baz alan model bize acaba dogru cevabi verebileccek mi. <br>dengesizlik, modeli öğrenirken taraflı (biased) hale getirir. Biz, upsampling veya downsampling yaparak, modeli "tarafsız" hale getiriyoruz. Modeli zorluyoruz ki, her iki durumu da ciddiye alsın ve sadece yüksek bir doğruluk oranı için değil, aynı zamanda gerçekten önemli olan olayları bulmak için de çaba göstersin.

```python
np.random.seed(42) #random degerlerde tutarlilik saglamasi icin. her calistirdigimizda . 42 popülerdir.

set1no=900
set2no=100
```

```python
data_frame1=pd.DataFrame({
    "feature_1" : np.random.normal(loc=0 , scale=1, size=set1no) ,#normal dagilima sahip olsun, 'loc'=mean, 'scale'=standart sapma, 'size'=boyut
    "feature_2" : np.random.normal(loc=0 , scale=1, size=set1no),
    "target" : [0]*set1no
})
```

```python
data_frame2=pd.DataFrame({
    "feature_1" : np.random.normal(loc=0 , scale=1, size=set2no) ,#normal dagilima sahip olsun, 'loc'=mean, 'scale'=standart sapma, 'size'=boyut
    "feature_2" : np.random.normal(loc=0 , scale=1, size=set2no),
    "target" : [1]*set2no
})
```

```python
print(data_frame1,"\n",data_frame2)
```

#### goruldugu uzere 900 tane 0, 100 tane 1 var . bunlari 'concat' yapabiliriz

```python
data_frame=pd.concat([data_frame1,data_frame2]).reset_index(drop=True)
```

```python
data_frame
```

```python
data_frame["target"].value_counts()# 0-->900, 1-->100
```

#### bu veriyi dengelemek icin bikac yol vardir.<br>1- upsampling--> dusuk sayida (minority)olan veriyi buyuk sayiya esitlemek. Yani 100 tane 1 degerini 900 tane yapmak<br>2-downsampling-->buyuk sayida(majortiy) olan veriyi kucultmek. Yani 900 tane olan 0 degerini 100'e indirgemek

### upsampling

```python
df_minority=data_frame[data_frame["target"] == 1]
df_majority=data_frame[data_frame["target"] == 0]
```

```python
print(df_minority,"\n",df_majority)
```

#### bu artirma olayini tabiki kafamiza gore yapmiyoruz. bu olay icin hazir olusturulmus kutuphaneleri kullanacagiz

```python
from sklearn.utils import resample #scikit-learn-->makine ogrenmesi icin kullanilir.cok kullancaz. cok ozelligi vardir.
```

```python
resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42) #n_samples=hangi sayiya cıkarmak istedigin
```

```python
df_minority_upsamples=resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42) #diyerek esitleyelim
```

```python
df_upsampled_concat=pd.concat([df_minority_upsamples, df_majority]).reset_index(drop=True) #diyerek concat edelim
```

```python
print(df_upsampled_concat,"\n")
print(df_upsampled_concat["target"].value_counts())
```

#### gorudugumuz gibi 1 ve 0 sayilari esitlenmis. Ama soyle bir durum var __upsampling yaparken eldeki datalari tekrarlayarak ekliyor__ . copy+paste gibi

### downsampled

```python
df_majority_downsampled=resample(df_majority, replace=True, n_samples=len(df_minority), random_state=42)
```

```python
df_majority_downsampled #oldusgunu gorecegizzz
```

```python
df_downsampled_concat=pd.concat([df_majority_downsampled, df_minority]).reset_index(drop=True)
```

```python
print(df_downsampled_concat,"\n")
print(df_downsampled_concat["target"].value_counts())
```

#### datalari upsampling yaparken tekrarlamak yerine benzerini eklemek icin bir algoritma gorcez bu cok daha mantiklidir aslinda. Daha iyi sonuclar alabiliriz

## SMOTE (Synthetic Minority Over-Sampling Technique)

```python
import matplotlib.pyplot as plt
plt.scatter(data_frame["feature_1"],data_frame["feature_2"] , c=data_frame["target"])
```

```python
from imblearn.over_sampling import SMOTE
```

#### imblearn --> imbalanced-learn yani dengesiz veriler icin olusturulmus

```python
oversample=SMOTE()
```

```python
(X,y)=oversample.fit_resample(data_frame[["feature_1","feature_2"]], data_frame["target"])
```

#### (X, y): y-->'target' yani hedef column'dur. Mesela bir ev icin fiyat ve bircok ozellik var. y=ev fiyatlarinin oldugu column olur.<br> X-->'feature' yani hedef column icin feature yan columnlar. Ev icin verdigimiz ornekte evin rengi, konumu, buyuklugu filan filan

```python
print(X,"\n",y)
```

#### gordugumuz gibi tam da oyle olmus, ve upsample yapmis. hepsinden de 1800 column var. bunlari 'concat' edebiliriz

```python
oversample_concat=pd.concat([X,y], axis=1) #columnlari concat ediyoruz
```

```python
print(oversample_concat,"\n",oversample_concat["target"].value_counts())
```

```python
plt.scatter(oversample_concat["feature_1"], oversample_concat["feature_2"], c=oversample_concat["target"])
```
