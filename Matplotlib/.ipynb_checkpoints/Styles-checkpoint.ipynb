```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
veri1=np.linspace(0,10,20)
print(veri1)
veri2=veri1**2
print(veri2)
```

#### plt.subplots() aslinda bize bir axes ve figure donduruyor. ve bunlari bir tuple olarak dondurur.-->(figure, axes) seklinde<br><br>__my_figure:__ Bu, tüm grafiğin ana kapsayıcısıdır. Tıpkı bir tuval gibi düşünebilirsin. Grafiğin boyutunu, başlığını ve genel ayarlarını buradan yaparsın. __my_axes:__ Bu, grafiğin kendisidir. Verilerin çizildiği, x ve y eksenlerinin, başlığın, etiketlerin olduğu asıl alandır. Sen bu my_axes nesnesini kullanarak çizimlerini yaparsın.

```python
my_figure, my_axes = plt.subplots()
my_axes.plot(veri1, veri2, color="#53ca0a") #color kismina rengin hexadecimal html bicimini koyabilin
my_axes.plot(veri2, veri1)
plt.show()
```

```python
new_fig, new_axes = plt.subplots()
new_axes.plot(veri1, veri1+2, color="blue", linewidth=1) #linewith-->kalinlik icib
new_axes.plot(veri1, veri1+3, color="yellow", linewidth=3)

new_axes.plot(veri1, veri1+4, color="darkblue", linewidth=1, linestyle="-.") #cizgileri degistirebiliyoz
new_axes.plot(veri1, veri1+5, color="darkblue", linewidth=1, linestyle=":")
new_axes.plot(veri1, veri1+6, color="darkblue", linewidth=1, linestyle="--")
new_axes.plot(veri1, veri1+7, color="black", linewidth=1, linestyle="--", marker="o", markersize=6, markerfacecolor="red") #pointleri dolgu yapabiliriz

plt.show()
```

## Farkli Grafikler

#### __-scatter plot__-->iste noktlarla, iki veri seti hakkindaki kolerasyonu olcerken filan kullanilir.<br> __-histogram__-->numeric degiskenlerin dagilimini gormek icin filan kullaniriz.<br> box plot, gibi gibi vardir.

### Scatter

```python
plt.scatter(veri1, veri2)
plt.show()
```

### Histogram

```python
newww=np.random.randint(0,100,20)
```

```python
plt.hist(newww)
plt.show()
```

### boxplot

```python
plt.boxplot(newww)
plt.show()
```
