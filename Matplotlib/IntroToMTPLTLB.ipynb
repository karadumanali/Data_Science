```python
import numpy as np
import matplotlib.pyplot as plt # import etmek icin
```

```python
%matplotlib inline 
```

##### #jupyder'de calisirken olustuedugumuz gorselleri  gorsellestirme icin bunu yazmamiz lazim matplotlib'de

```python
yas_listesi=[15,18,20,22,33,40,47,55]
kilo_listesi=[54,60,60,70,84,65,70,61]
```

```python
plt.plot(yas_listesi, kilo_listesi, "r") #bu yaptigi islemi hafizaya kaydeder., 'r'-->red, 'b'-->blue ...
plt.xlabel("yas") #x-ekseninin adi
plt.ylabel("kilo")
plt.title("yas-kilo") #baslik
plt.show() #'da goster demek
#plt.savefig("grafik.jpg") #dersek bulundugumuz konuma grafigin resmini kaydedecektir
```

##### mesela bu kodu sunucuda vs calistirirsak goremeyebiliriz demi

## numpy ile de calisabiliriz

```python
np_yas_listesi=np.array(yas_listesi)
np_kilo_listesi=np.array(kilo_listesi)
```

```python
plt.plot(np_yas_listesi, np_kilo_listesi, "grey")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("yas-kilo")
plt.show()
```

##### goruldugu uzere listeyle ayni grafigi verdi onu göstermek icin yaptik.

```python
np_arr1=np.linspace(0,10,20) #0 ve 10 dahil arasini 20 esit parcaya boluyor. yani her eleman arasi mesafe esit oluyor.
```

```python
print(np_arr1) #goruldugu gibi
```

##### burda üstel fonksiyon grafigi denicez yani __np_arr1'in__ ustel olarak cizilmesini saglicaz

```python
np_arr2=np_arr1 ** 3
print(np_arr2)
```

```python
plt.plot(np_arr1, np_arr2, "b*-") #"*"-->*'li cikmasini saglar pointlerin,"-" --> duz cizgi seklinde,"--"--> ayni kendi seklinde kesik kesik
plt.show()
```

##### biz np_arr1 ve np_arr2'yi ayri ayri cizmek isteyebiliriz ayni kod icinde

```python
plt.subplot(1,2,1) #-->ilk 1-bir tane row olsun, -->2: iki tane column olsun, -->son 1- birinci grafigi ciziyorum
plt.plot(np_arr1, np_arr2, "g*-")
plt.subplot(1,2,2)
plt.plot(np_arr2, np_arr1, "b--") 
plt.show()
```

##### bir row=ikisi yan yana, iki column=iki tane grafik var<br> tabi eksenleri kendisi olusturuyor burda mantiken

##### biz simdi figure olusturucaz. figuru olusturmamizin nedeni icerisine kendi eksenlerimizi ve onlarin adini filan kendimiz koyabiliyoruz<br>biz surekli plt.show() diyip duruyoruz ama o an hangisi ciziliyse onu cizer o ve pycharm gibi seylerde sikinti cikarabilir. Jupyterde sirayla gittigimiz icin sikinti yok

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.4,0.4]) #0.1 ler onemli degil. ilk 0.4-->genislik, ikinci 0.4-->yukseklik
figure_axes.plot(np_arr1, np_arr2,"grey")
figure_axes.set_xlabel("x-ekseni")
figure_axes.set_ylabel("y-ekseni")
figure_axes.set_title("baslik")
plt.show()
```

##### biz burda kendi grafigimizi olusturduk

```python
new_fig=plt.figure(dpi=100) #dpi cozunurlugu verebiliriz
new_axes=new_fig.add_axes([0.1,0.1,0.6,0.6])
new_axes.plot(np_arr1, np_arr1 ** 2 , label="np_arr1 ** 2")
new_axes.plot(np_arr1, np_arr1 ** 3 , label="np_arr1 ** 3")

new_axes.legend() #label'lerin gozukmesi icin. legend(loc=x) x yerine 1,2,3,4 .. gibi degerler girerek nerde duracagini belirtebilriiz
plt.show()
```
